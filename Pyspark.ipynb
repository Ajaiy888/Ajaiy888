```python
from pyspark.sql import SparkSession
from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
# Define the data
data = [Row(1, "John", 30, "Sales", 50000.0),
 Row(2, "Alice", 28, "Marketing", 60000.0),
 Row(3, "Bob", 32, "Finance", 55000.0),
 Row(4, "Sarah", 29, "Sales", 52000.0),
 Row(5, "Mike", 31, "Finance", 58000.0)
]
```

```python
schema = StructType([
 StructField("id", IntegerType(), nullable=False),
 StructField("name", StringType(), nullable=False),
 StructField("age", IntegerType(), nullable=False),
 StructField("department", StringType(), nullable=False),
 StructField("salary", DoubleType(), nullable=False)
])
```

```python
employeeDF = spark.createDataFrame(data, schema)
```

```python
employeeDF.show()
```

```python
#Calculate the average salary for each department:
from pyspark.sql.functions import avg

averagesalbydept = (
    employeeDF.groupBy("department")
    .agg(
        avg("salary").alias("avg_salary")
    )
)
averagesalbydept.show()
```

```python
#add a new column named "bonus" thaat is 10% of the salary for all employees
employeeDF = employeeDF.selectExpr('*', "salary * 0.1 as bonus")
employeeDF.show()
```

```python
#add a new column named "bonus" thaat is 10% of the salary for all employees
employeeDF = employeeDF.withColumn("bonus", col("salary") * 0.1)
employeeDF.show()
```

```python
#Group the data by department and find the employee with
#the highest salary in each department
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

windowSpec = Window.partitionBy("department").orderBy("salary")
highestSalByDept = employeeDF.withColumn("row_number", row_number().over(windowSpec)).filter(col("row_number") == 1).drop("row_number")
highestSalByDept.show()
```

```python
#Find the top 3 departments with the highest total salary
from pyspark.sql.functions import sum

top3Dept = (
    employeeDF.groupBy("department")
    .agg(
        sum("salary").alias("total_salary")
    )
    .orderBy(col("total_salary").desc())
    .limit(3)
)
top3Dept.show()
```

```python
#find the top most department having highest salary
from pyspark.sql.functions import sum

topDept = (
    employeeDF.groupBy("department")
    .agg(
        sum("salary").alias("total_salary")
    )
    .orderBy(col("total_salary").desc())
    .limit(1)
)
topDept.show()
```

```python
#Find the top most department having highest salary
employee_CTE = employeeDF.groupBy("department").agg(sum("salary").alias("total_salary"))
windowSpec = Window.orderBy(col("total_salary").desc())
department_rank = employee_CTE.withColumn("row_number", row_number().over(windowSpec))
topDept = department_rank.filter(col("row_number") == 1).select("department")
topDept.show()
```

```python
#Filter the DataFrame to keep only employees aged 30 or above and
#working in the "Sales" department
above30Sales = employeeDF.filter((col("age") >= 30) & (col("department") == "Sales"))
above30Sales.show()
```

```python
#Calculate the difference between each employee's salary and the average
#salary of their respective department
windowSpec = Window.partitionBy("department")
employeeDF = employeeDF.withColumn("avg_salary", avg("salary").over(windowSpec))
from pyspark.sql.functions import col, expr

diffSalDf = employeeDF.withColumn("salary_diff", expr("salary - avg_salary"))
diffSalDf.show()
```

```python
#Calculate the sum of salaries for employees whose names start with the letter "J"
sumSalaries = employeeDF.filter(col("name").startswith("J")).agg(sum("salary").alias("sum_salary"))
sumSalaries.show
```
