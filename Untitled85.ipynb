```python
import tensorflow as tf

# Scalars, vectors, matrices
a = tf.constant(5)
b = tf.constant([1,2,3])
c = tf.constant([[1,2],[3,4]])

print(a)
print(b)
print(c)
```

```python
x = tf.constant([1,2,3])
y = tf.constant([10,20,30])

print(x + y)
print(x * y)
print(tf.matmul([[1,2]], [[2],[3]]))  # matrix multiply
```

```python
v = tf.Variable(3.0)
print(v)

v.assign(10.0)
print(v)
```

```python
import numpy as np
import tensorflow as tf

X = np.array([1,2,3,4], dtype=float)
y = np.array([4,7,10,13], dtype=float)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])

model.compile(optimizer='sgd', loss='mse')

model.fit(X, y, epochs=200)
```

```python
print(model.predict(np.array([10])))  # should be around 31
```
